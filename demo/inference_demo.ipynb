```python
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
import mmcv
```

```python
config_file = '/home/zhongying/research/repo/mmdetection/configs/polarmask/polar_768_1x_r50.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = '/home/zhongying/research/repo/mmdetection/work_dirs/polar_768_1x_r50/epoch_12.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
img = '/home/zhongying/research/repo/mmdetection/demo/demo.jpg'
result = inference_detector(model, img)
```
